```python
import os

def print_folder_structure(folder_path):
    for root, dirs, files in os.walk(folder_path):
        level = root.replace(folder_path, '').count(os.sep)
        indent = ' ' * 4 * level
        print(f"{indent}{os.path.basename(root)}/")
        sub_indent = ' ' * 4 * (level + 1)
        for file in files:
            print(f"{sub_indent}{file}")

# Пример использования
print_folder_structure('/home/belyaeva.a/пациенты и контроль')
```

```python
import os

def sanitize_name(name):
    """
    Удаляет круглые скобки и цифры внутри них из имени файла или папки.
    """
    import re
    return re.sub(r'\(\d+\)', '', name).replace('()', '').strip()

def sanitize_directory(base_dir):
    """
    Проходит по всем файлам и папкам в указанной директории
    и переименовывает их, удаляя `()` и цифры внутри.
    """
    for dirpath, dirnames, filenames in os.walk(base_dir, topdown=False):
        # Обрабатываем файлы
        for file in filenames:
            old_path = os.path.join(dirpath, file)
            new_name = sanitize_name(file)
            new_path = os.path.join(dirpath, new_name)
            if old_path != new_path:
                os.rename(old_path, new_path)

        # Обрабатываем папки
        for dirname in dirnames:
            old_path = os.path.join(dirpath, dirname)
            new_name = sanitize_name(dirname)
            new_path = os.path.join(dirpath, new_name)
            if old_path != new_path:
                os.rename(old_path, new_path)

# Укажите путь к папке, которую хотите обработать
base_directory = '/home/belyaeva.a/пациенты и контроль'
sanitize_directory(base_directory)

print(f"Sanitization completed for '{base_directory}'.")
```

```python
import mne
import os

# Define the path to your data folder
data_dir = "/home/belyaeva.a/пациенты и контроль/open_eyes"

# Selected examples with their labels
examples = [
    {"patient_id": "KARISHEVA_ADELINA_49_EO_free", "label": "control"},
    {"patient_id": "Belkin Alex_20_f42-2_ОГ", "label": "ОКР"},
    {"patient_id": "PoIP_f22_f60-3_f41-2_Na_As_Ep_ог", "label": "ПРЛ"},
    {"patient_id": "VaAA_f22_f41-2_f32-1_og", "label": "тревожно-фобические"}
]

# Define the 10-20 Electrode System
channels_to_use = [
    'Fp1', 'Fp2', 'F7', 'F3', 'Fz', 'F4', 'F8', 'T3', 'C3', 'Cz', 'C4', 
    'T4', 'T5', 'P3', 'Pz', 'P4', 'T6', 'O1', 'O2'
]

# Convert channels_to_use to lowercase for consistency
channels_to_use = [ch.lower() for ch in channels_to_use]

# Function to process each EDF file
def process_edf(patient_id, label):
    file_path = os.path.join(data_dir, label, f"{patient_id}.edf")
    try:
        # Load the EDF file
        raw = mne.io.read_raw_edf(file_path, preload=True, verbose=False)

        # Step 1: Standardize channel names
        raw.rename_channels({ch: ch.replace(".", "").lower() for ch in raw.ch_names})

        # Step 2: Print available channels for debugging
        print(f"Available channels in {patient_id}: {raw.ch_names}")

        # Step 3: Get the 10-20 montage
        ten_twenty_montage = mne.channels.make_standard_montage("standard_1020")

        # Step 4: Find matching channels
        matching_channels = [ch for ch in channels_to_use if ch in raw.ch_names]
        if not matching_channels:
            print(f"No matching channels found for {patient_id}. Skipping...")
            return None

        # Step 5: Pick only matching channels
        raw.pick_channels(matching_channels)

        # Step 6: Apply the montage with on_missing="ignore"
        raw.set_montage(ten_twenty_montage, on_missing="ignore")

        # Verify the montage has been set correctly
        print(f"Processed {patient_id} ({label}):")
        print(raw.info)

        return raw

    except Exception as e:
        print(f"Error processing {patient_id} ({label}): {e}")
        return None

# Process all examples
processed_samples = []
for example in examples:
    raw_data = process_edf(example["patient_id"], example["label"])
    if raw_data:
        processed_samples.append(raw_data)
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Define frequency bands with corresponding colors
bands = {
    "delta": (1, 4, "blue"),
    "theta": (4, 8, "green"),
    "alpha": (8, 13, "yellow"),
    "beta": (13, 30, "orange"),
    "gamma": (30, 45, "red")
}

# Function to plot PSD for a given raw object
def plot_psd_by_label(raw, label, bands, ax=None):
    """
    Plot the PSD for the given raw object and label.

    Parameters:
    - raw: MNE Raw object
    - label: Label of the subject (e.g., "ОКР")
    - bands: Dictionary of frequency bands
    - ax: Matplotlib axis (optional)
    """
    psds, freqs = mne.time_frequency.psd_welch(raw, fmin=0.5, fmax=50, n_fft=1024, verbose=False)
    psds = 10 * np.log10(psds)  # Convert to dB

    if ax is None:
        ax = plt.gca()
    
    # Plot the PSD for each channel
    for idx, ch_name in enumerate(raw.ch_names):
        ax.plot(freqs, psds[idx], label=ch_name, linewidth=0.7)

    # Highlight frequency bands with different colors
    for band, (low, high, color) in bands.items():
        ax.axvspan(low, high, color=color, alpha=0.2, label=f"{band} band")

    ax.set_title(f"PSD for {label}")
    ax.set_xlabel("Frequency (Hz)")
    ax.set_ylabel("Power Spectral Density (dB)")
    ax.legend(loc="upper right", fontsize="small")

# Visualize PSDs for each label
fig, axs = plt.subplots(len(processed_samples), 1, figsize=(10, 5 * len(processed_samples)))

for i, raw in enumerate(processed_samples):
    label = examples[i]["label"]  # Retrieve the label
    plot_psd_by_label(raw, label, bands, ax=axs[i])

plt.tight_layout()
plt.show()

```

```python
from scipy import signal

def get_filter(sfreq=250., band='alpha'):
    """
    Generate a FIR filter for the specified frequency band.

    Parameters:
    - sfreq (float): Sampling frequency of the signal.
    - band (str): Frequency band name as defined in band_bounds.

    Returns:
    - filt (numpy array): FIR filter coefficients.
    """
    # Validate if the band exists in band_bounds
    if band not in band_bounds:
        raise ValueError(f"Band '{band}' is not defined in band_bounds.")

    # Extract band bounds
    f_low = band_bounds[band][0]
    f_high = band_bounds[band][1]

    # Define filter transition regions (add a margin for smoothness)
    f_low_lb = max(0, f_low - 1)  # Ensure frequencies are non-negative
    f_low_ub = f_low
    f_high_lb = f_high
    f_high_ub = f_high + 1

    nyq = sfreq / 2.  # The Nyquist frequency (half the sampling rate)

    # Define frequency and gain profiles for the FIR filter
    freq = [0., f_low_lb, f_low_ub, f_high_lb, f_high_ub, nyq]
    gain = [0, 0, 1, 1, 0, 0]

    # Length of the filter (1 second worth of samples)
    n = int(round(sfreq)) + 1  # Ensures odd filter length for zero-phase filtering

    # Generate FIR filter coefficients
    filt = signal.firwin2(n, freq, gain, nyq=nyq)

    return filt
```

```python
import numpy as np
import pandas as pd
from scipy.signal import welch, hilbert
from scipy.stats import pearsonr, entropy
from mne.connectivity import spectral_connectivity
from mne.time_frequency import psd_array_welch
from scipy import signal
from itertools import combinations
from joblib import Parallel, delayed
from scipy.linalg import svd
import antropy as ant
from nolds import dfa, lyap_r, hurst_rs
import random

random.seed(42)
np.random.seed(42)

# Precompute filters for each band
filters = {band: signal.firwin2(
    numtaps=int(round(125)) + 1,
    freq=[0, max(0, low - 1), low, high, high + 1, 125 / 2],
    gain=[0, 0, 1, 1, 0, 0],
    nyq=125 / 2
) for band, (low, high) in band_bounds.items()}


def compute_band_power(signal_array, sfreq, band_bounds):
    """Compute power for all bands in a single pass."""
    freqs, psd = welch(signal_array, fs=sfreq, nperseg=min(256, len(signal_array)))
    total_power = psd.sum()
    band_powers = {band: psd[(freqs >= low) & (freqs <= high)].sum() / total_power
                   for band, (low, high) in band_bounds.items()}
    return band_powers


def compute_envelope_features(signal_array, filt, band):
    """Compute envelope standard deviation and correlation."""
    filtered_signal = np.convolve(filt, signal_array, 'valid')
    envelope = np.abs(hilbert(filtered_signal))
    std_ratio = envelope.std() / envelope.mean() if envelope.mean() > 0 else 0
    return std_ratio, envelope


def compute_non_linear_features(signal_array, sfreq):
    """Compute non-linear features for the signal."""
    sanitized_signal = np.array(signal_array, dtype=np.float64)
    sanitized_signal = sanitized_signal[np.isfinite(sanitized_signal)]  # Remove NaN/Inf
    if len(sanitized_signal) < 10:
        return {key: np.nan for key in [
            'approx_entropy', 'sample_entropy', 'spectral_entropy', 'higuchi_fd',
            'dfa','hurst_exp', 'diagonal_entropy', 'fisher_info'
        ]}

    non_linear_feats = {}

    # (a) Entropy measures
    non_linear_feats['approx_entropy'] = ant.app_entropy(sanitized_signal)
    non_linear_feats['sample_entropy'] = ant.sample_entropy(sanitized_signal)
    non_linear_feats['spectral_entropy'] = ant.spectral_entropy(sanitized_signal, sfreq, method='welch')

    # (b) Fractal dimensions
    non_linear_feats['higuchi_fd'] = ant.higuchi_fd(sanitized_signal)

    # (c) DFA (Detrended Fluctuation Analysis)
    non_linear_feats['dfa'] = dfa(sanitized_signal)

    # (e) Hurst exponent
    non_linear_feats['hurst_exp'] = hurst_rs(sanitized_signal)

    # (g) Recurrence analysis features
    recurrence_matrix = np.abs(sanitized_signal[:, None] - sanitized_signal) < 0.1  # Threshold example
    diagonal_lengths = [np.sum(np.diag(recurrence_matrix, k)) for k in range(-len(sanitized_signal), len(sanitized_signal))]
    non_linear_feats['diagonal_entropy'] = entropy(diagonal_lengths)

    # (h) Chaos measures
    non_linear_feats['fisher_info'] = np.var(sanitized_signal) / np.mean(sanitized_signal**2)

    return non_linear_feats

def compute_additional_features(signals, channel_names, sfreq):
    """Вычисляет дополнительные признаки напрямую из сигналов."""
    features = {}
    
    # Низкочастотные мощности для каждого канала
    for idx, ch_name in enumerate(channel_names):
        signal = signals[idx]
        freqs, psd = welch(signal, fs=sfreq, nperseg=min(256, len(signal)))
        total = psd.sum()
        for band in ['delta', 'theta']:
            low, high = band_bounds[band]
            features[f'{ch_name}_{band}_rel_power'] = psd[(freqs >= low) & (freqs <= high)].sum() / total
    
    # Альфа микросостояния
    try:
        gfp = signals.std(axis=0)
        peaks = signal.find_peaks(gfp, height=np.percentile(gfp, 95))[0]
        maps = signals[:, peaks].T
        labels = KMeans(n_states=4, random_state=0).fit(maps).predict(signals.T)
        
        for s in range(4):
            sig = signals[:, labels==s].mean(axis=1) if np.any(labels==s) else np.zeros(signals.shape[0])
            freqs, psd = welch(sig, fs=sfreq, nperseg=min(256, len(sig)))
            low, high = band_bounds['alpha']
            features[f"ms{s}_alpha_power"] = psd[(freqs >= low) & (freqs <= high)].sum()
            
        # Временные характеристики микросостояний
        for s in range(4):
            runs = np.diff(np.where(np.concatenate(([labels[0]],
                                                   labels[:-1]!=labels[1:],
                                                   [True])))[0])[::2]
            features[f"ms{s}_dur"] = runs.mean()/sfreq
            features[f"ms{s}_occ"] = len(runs)/(len(labels)/sfreq)
            features[f"ms{s}_cov"] = (labels==s).mean()
        
        # Нелинейные характеристики микросостояний
        features['ms_sample_entropy'] = ant.sample_entropy(labels)
        features['ms_lz_complexity'] = ant.lziv_complexity(labels, normalize=True)
        features['ms_hurst_index'] = hurst_rs(labels)
    except Exception as e:
        print(f"Ошибка при вычислении микросостояний: {str(e)}")
    
    # Бета мощность
    for idx, ch_name in enumerate(channel_names):
        signal = signals[idx]
        freqs, psd = welch(signal, fs=sfreq, nperseg=min(256, len(signal)))
        beta_mask = (freqs >= 13) & (freqs <= 30)
        features[f'{ch_name}_beta_power'] = psd[beta_mask].mean()
    
    # Отношение тета/бета
    for idx, ch_name in enumerate(channel_names):
        signal = signals[idx]
        freqs, psd = welch(signal, fs=sfreq, nperseg=min(256, len(signal)))
        total = psd.sum()
        theta_power = psd[(freqs >= band_bounds['theta'][0]) & (freqs <= band_bounds['theta'][1])].sum() / total
        beta_power = psd[(freqs >= 13) & (freqs <= 30)].sum() / total
        features[f'{ch_name}_theta_beta_ratio'] = theta_power / beta_power if beta_power > 0 else 0
    
    # Когерентность между парами каналов
    for i, j in combinations(range(len(channel_names)), 2):
        ch1, ch2 = channel_names[i], channel_names[j]
        sig1, sig2 = signals[i], signals[j]
        
        # Бета-диапазон когерентность
        freqs, csd = signal.csd(sig1, sig2, fs=sfreq, nperseg=min(256, len(sig1)))
        freqs, psd1 = welch(sig1, fs=sfreq, nperseg=min(256, len(sig1)))
        _, psd2 = welch(sig2, fs=sfreq, nperseg=min(256, len(sig2)))
        
        beta_mask = (freqs >= 13) & (freqs <= 30)
        coherence = np.abs(csd[beta_mask])**2 / (psd1[beta_mask] * psd2[beta_mask])
        features[f'coh_beta_{ch1}_{ch2}'] = np.mean(coherence)
        
        # Гамма-диапазон когерентность
        gamma_mask = (freqs >= 30) & (freqs <= 100)
        coherence = np.abs(csd[gamma_mask])**2 / (psd1[gamma_mask] * psd2[gamma_mask])
        features[f'coh_gamma_{ch1}_{ch2}'] = np.mean(coherence)
    
    # Мультимасштабная энтропия и относительная мощность
    for idx, ch_name in enumerate(channel_names):
        signal = signals[idx]
        try:
            mse = ant.multiscale_entropy(signal, scale=5)
            features[f'{ch_name}_multiscale_entropy'] = np.mean(mse)
            
            freqs, psd = welch(signal, fs=sfreq, nperseg=min(256, len(signal)))
            total = psd.sum()
            for band in ['theta', 'alpha']:
                low, high = band_bounds[band]
                features[f'{ch_name}_rel_{band}'] = psd[(freqs >= low) & (freqs <= high)].sum() / total
        except Exception as e:
            print(f"Ошибка при вычислении MSE для канала {ch_name}: {str(e)}")
    
    # Относительная спектральная мощность
    for idx, ch_name in enumerate(channel_names):
        signal = signals[idx]
        freqs, psd = welch(signal, fs=sfreq, nperseg=min(256, len(signal)))
        total = psd.sum()
        for band, (low, high) in band_bounds.items():
            if band != 'gamma':
                band_power = psd[(freqs >= low) & (freqs <= high)].sum()
                features[f'{ch_name}_rel_{band}'] = band_power / total
    
    # Вейвлет энергии
    try:
        all_e = []
        for signal in signals:
            coeffs = pywt.wavedec(signal, 'db4', level=5)
            all_e.append([np.sum(c**2) for c in coeffs[1:]])
        avg = np.mean(all_e, axis=0)
        for j in range(5):
            features[f"wavelet_E{j+1}"] = avg[j]
    except Exception as e:
        print(f"Ошибка при вычислении вейвлет-энергий: {str(e)}")
    return features

def extract_features(row, channels, sfreq):
    """Extract features for a single row."""
    feats = {'patient_id': row['patient_id'], 'label': row['label']}
    channel_data = {ch: row[ch] for ch in channels if isinstance(row[ch], list) and len(row[ch]) > 0}
    channel_signals = np.array(list(channel_data.values()))
    
    # Существующие признаки
    for idx, channel in enumerate(channel_data):
        band_powers = compute_band_power(channel_signals[idx], sfreq, band_bounds)
        for band, power in band_powers.items():
            feats[f'{channel}_{band}_power'] = power

    for band, filt in filters.items():
        envelopes = []
        for idx, signal_array in enumerate(channel_signals):
            std_ratio, envelope = compute_envelope_features(signal_array, filt, band)
            feats[f'{list(channel_data.keys())[idx]}_env_std_{band}'] = std_ratio
            envelopes.append(envelope)

        for i, j in combinations(range(len(envelopes)), 2):
            ch1 = list(channel_data.keys())[i]
            ch2 = list(channel_data.keys())[j]
            feats[f'env_corr_{ch1}_{ch2}_{band}'] = pearsonr(envelopes[i], envelopes[j])[0]

    for idx, signal_array in enumerate(channel_signals):
        non_linear_feats = compute_non_linear_features(signal_array, sfreq)
        for feat_name, feat_value in non_linear_feats.items():
            feats[f'{list(channel_data.keys())[idx]}_{feat_name}'] = feat_value     
    # Добавляем новые признаки
    try:
        additional_features = compute_additional_features(
            channel_signals, 
            list(channel_data.keys()), 
            sfreq
        )
        feats.update(additional_features)
    except Exception as e:
        print(f"Ошибка при вычислении дополнительных признаков: {str(e)}")

    return feats

# Extract features for all rows in parallel
channels = ['C3', 'C4', 'Cz', 'F3', 'F4', 'F7', 'F8', 'Fp2', 'Fz', 'P3', 'P4', 'Pz', 'T3', 'T4']
sfreq = 250.  # Replace with actual sampling frequency if variable
all_feats = Parallel(n_jobs=-1)(delayed(extract_features)(row, channels, sfreq) for _, row in df.iterrows())

# Create a DataFrame with extracted features
feats_df = pd.DataFrame(all_feats)
```

```python
import shap
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.model_selection import StratifiedKFold, train_test_split
from sklearn.metrics import f1_score
import numpy as np
import pandas as pd
import random

# Фиксируем случайное состояние для воспроизводимости
random.seed(42)
np.random.seed(42)


# 1. Разбивка данных на train+validation и test (80% и 20% соответственно)
X_train_val, X_test, y_train_val, y_test = train_test_split(
    X, y, test_size=0.2, stratify=y, random_state=42
)

# 2. Из train+validation выделяем отдельную валидационную выборку (25% от train_val)
# Итоговое соотношение: 60% train, 20% validation, 20% test
X_train, X_val, y_train, y_val = train_test_split(
    X_train_val, y_train_val, test_size=0.25, stratify=y_train_val, random_state=42
)

# 3. Инициализация моделей
models = {
    "RandomForest": RandomForestClassifier(class_weight="balanced", random_state=42),
    "LogisticRegression": LogisticRegression(class_weight="balanced", max_iter=1000, random_state=42),
    "SVM": SVC(class_weight="balanced", probability=True, random_state=42)
}

# 4. Инициализация кросс-валидации на обучающей выборке
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
results = {}

# Словари для хранения SHAP-значений и важности признаков (на основе CV)
shap_values_cv = {}
feature_importances = {}

# Базовая модель: предсказание наиболее часто встречающегося класса на train
f1_baseline_scores = []
for train_idx, test_idx in skf.split(X_train, y_train):
    y_train_cv = y_train.iloc[train_idx]
    y_test_cv = y_train.iloc[test_idx]
    most_common_class = y_train_cv.value_counts().idxmax()
    y_pred_baseline = np.full_like(y_test_cv, most_common_class)
    f1 = f1_score(y_test_cv, y_pred_baseline, average="macro")
    f1_baseline_scores.append(f1)
results["MostCommon"] = {
    "mean": np.mean(f1_baseline_scores),
    "std": np.std(f1_baseline_scores)
}

# 5. Кросс-валидация на обучающей выборке для оценки моделей и сбора SHAP-значений
for model_name, model in models.items():
    f1_scores_cv = []
    shap_vals_list = []  # для хранения SHAP-значений с каждого fold

    for train_idx, test_idx in skf.split(X_train, y_train):
        # Разбиваем train на CV-обучающую и CV-тестовую выборки
        X_train_cv, X_test_cv = X_train.iloc[train_idx], X_train.iloc[test_idx]
        y_train_cv, y_test_cv = y_train.iloc[train_idx], y_train.iloc[test_idx]

        # Обучаем модель и делаем предсказание
        model.fit(X_train_cv, y_train_cv)
        y_pred = model.predict(X_test_cv)
        f1 = f1_score(y_test_cv, y_pred, average="macro")
        f1_scores_cv.append(f1)

        # SHAP-анализ
        if model_name == "SVM":
            explainer = shap.KernelExplainer(model.predict_proba, X_train_cv)
            shap_vals = explainer.shap_values(X_test_cv)
            # If you're sure it's binary, shap_vals might be a list with 2 elements
            # shap_vals = shap_vals[1] if isinstance(shap_vals, list) else shap_vals
            
            # Check shape; if needed, transpose
            if isinstance(shap_vals, list):
                # For binary classification, pick the second class
                shap_vals = shap_vals[1]
            if shap_vals.shape != (X_test_cv.shape[0], X_test_cv.shape[1]):
                shap_vals = shap_vals.T

        elif model_name == "LogisticRegression":
            explainer = shap.LinearExplainer(model, X_train_cv)
            shap_vals = explainer.shap_values(X_test_cv)
            # If multi-class, shap_vals might be a list; for binary, it's often an array
            if isinstance(shap_vals, list):
                # For binary classification, pick the second class
                shap_vals = shap_vals[1]
            # Transpose if shape is (n_features, n_samples) instead of (n_samples, n_features)
            if shap_vals.shape != (X_test_cv.shape[0], X_test_cv.shape[1]):
                shap_vals = shap_vals.T

        else:  # RandomForest
            explainer = shap.TreeExplainer(model)
            shap_vals = explainer.shap_values(X_test_cv)
            if isinstance(shap_vals, list):
                shap_vals = shap_vals[1]
            # Usually TreeExplainer returns shape (n_samples, n_features) for single-class

        shap_vals_list.append(shap_vals)

    # Сохраняем средний F1 score
    results[model_name] = {
        "mean": np.mean(f1_scores_cv),
        "std": np.std(f1_scores_cv)
    }
    # Объединяем SHAP-значения из всех fold
    shap_values_all = np.concatenate(shap_vals_list, axis=0)
    # Вычисляем важность признаков (средний абсолютный вклад)
    feature_importances[model_name] = np.abs(shap_values_all).mean(axis=0)
    # Если хотите сохранить shap_values_all для дальнейшего анализа:
    shap_values_cv[model_name] = shap_values_all

# 6. Вывод результатов моделей (оценка на обучении через CV)
print("\nModel Results on Training Set (via CV):")
for model_name, metrics in results.items():
    print(f"{model_name}: {metrics['mean']:.4f} ± {metrics['std']:.4f}")

# 7. Оценка моделей на валидационной выборке
print("\nValidation Set Evaluation:")
for model_name, model in models.items():
    model.fit(X_train, y_train)
    y_val_pred = model.predict(X_val)
    f1_val = f1_score(y_val, y_val_pred, average="macro")
    print(f"{model_name} on validation: F1 = {f1_val:.4f}")

# 8. Финальная оценка на тестовой выборке и построение SHAP summary plot
print("\nTest Set Evaluation and SHAP Summary Plot:")
for model_name, model in models.items():
    # Обучаем модель на объединённых train+validation данных
    model.fit(X_train_val, y_train_val)
    y_test_pred = model.predict(X_test)
    f1_test = f1_score(y_test, y_test_pred, average="macro")
    print(f"{model_name} on test: F1 = {f1_test:.4f}")

    # Вычисляем SHAP-значения для тестовой выборки
    if model_name == "SVM":
        explainer = shap.KernelExplainer(model.predict_proba, X_train_val)
        shap_vals_test = explainer.shap_values(X_test)
        if isinstance(shap_vals_test, list):
            shap_vals_test = shap_vals_test[1]
        if shap_vals_test.shape != (X_test.shape[0], X_test.shape[1]):
            shap_vals_test = shap_vals_test.T

    elif model_name == "LogisticRegression":
        explainer = shap.LinearExplainer(model, X_train_val)
        shap_vals_test = explainer.shap_values(X_test)
        if isinstance(shap_vals_test, list):
            shap_vals_test = shap_vals_test[1]
        if shap_vals_test.shape != (X_test.shape[0], X_test.shape[1]):
            shap_vals_test = shap_vals_test.T

    else:  # RandomForest
        explainer = shap.TreeExplainer(model)
        shap_vals_test = explainer.shap_values(X_test)
        if isinstance(shap_vals_test, list):
            shap_vals_test = shap_vals_test[1]

    print(f"\nSHAP Summary Plot for {model_name}:")
    shap.summary_plot(shap_vals_test, X_test, feature_names=X_test.columns)

# 9. Анализ топ-N признаков по важности для каждой модели (на основе CV)
top_n = 10
for model_name, importances in feature_importances.items():
    print(f"\nTop {top_n} features for {model_name}:")
    sorted_indices = np.argsort(importances)[::-1][:top_n]
    for idx in sorted_indices:
        print(f"{X.columns[idx]}: {importances[idx]:.4f}")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import welch
import ast

# Загрузка DataFrame
open_eyes_df = pd.read_csv('/home/belyaeva.a/data_open.csv')
closed_eyes_df = pd.read_csv('/home/belyaeva.a/data_closed.csv')

# Определение ритмов
rhythms = {
    "delta": (0.5, 4),
    "theta": (4, 8),
    "alpha": (8, 12),
    "beta": (12, 30),
    "mu": (8, 13)
}

# Каналы и топологические группы
channels = {
    "central": ['C3', 'C4', 'Cz'],
    "frontal": ['F3', 'F4', 'Fz'],
    "parietal": ['P3', 'P4', 'Pz'],
    "occipital": ['O1', 'O2']
}

# Функция для преобразования строк в списки
def convert_to_list(value):
    if isinstance(value, str):
        try:
            # Попытка преобразования строки в список
            converted = ast.literal_eval(value)
            if isinstance(converted, list) and all(isinstance(x, (int, float)) for x in converted):
                return converted
        except (ValueError, SyntaxError):
            return None
    elif isinstance(value, float):
        return [value] if not pd.isna(value) else None
    return value

    
# Функция для нормализации patient_id
def normalize_patient_id(patient_id):
    # Список суффиксов для удаления
    suffixes_to_remove = ['_ЗГ', '_zg', '_ec', '_og', '_EO', '_ОГ']
    lower_patient_id = str(patient_id).strip().lower()
    for suffix in suffixes_to_remove:
        if lower_patient_id.endswith(suffix.lower()):
            return lower_patient_id[: -len(suffix)]
    return lower_patient_id

# Нормализация patient_id в обеих таблицах
open_eyes_df['patient_id'] = open_eyes_df['patient_id'].apply(normalize_patient_id)
closed_eyes_df['patient_id'] = closed_eyes_df['patient_id'].apply(normalize_patient_id)

# Преобразование данных в списки для всех каналов
for channel in sum(channels.values(), []):
    open_eyes_df[channel] = open_eyes_df[channel].apply(convert_to_list)
    closed_eyes_df[channel] = closed_eyes_df[channel].apply(convert_to_list)

# Найти общие patient_id
common_ids = set(open_eyes_df['patient_id']).intersection(set(closed_eyes_df['patient_id']))
print(f"Common patient IDs: {common_ids}")

# Фильтрация данных по общим patient_id
open_eyes_df = open_eyes_df[open_eyes_df['patient_id'].isin(common_ids)]
closed_eyes_df = closed_eyes_df[closed_eyes_df['patient_id'].isin(common_ids)]

# Функция для вычисления PSD
def compute_psd(data, sfreq, rhythm_ranges, channels_group):
    psd_results = {rhythm: {group: [] for group in channels_group} for rhythm in rhythm_ranges}
    for group, group_channels in channels_group.items():
        for channel in group_channels:
            if channel in data.columns and isinstance(data[channel].iloc[0], list):
                channel_data = np.array(data[channel].iloc[0])  # Данные только для одного канала
                freqs, psd = welch(channel_data, sfreq, nperseg=256)
                for rhythm, (low, high) in rhythm_ranges.items():
                    idx = (freqs >= low) & (freqs < high)
                    if np.any(idx):  # Проверка, что индексы не пустые
                        psd_results[rhythm][group].append(np.mean(psd[idx]))
                    else:
                        psd_results[rhythm][group].append(np.nan)  # Если индексы пустые, добавляем NaN
    # Усреднение значений по каждой группе
    for rhythm in psd_results:
        for group in psd_results[rhythm]:
            psd_results[rhythm][group] = np.mean(psd_results[rhythm][group]) if psd_results[rhythm][group] else np.nan
    return pd.DataFrame(psd_results).T

# Сравнение топологии между open_eyes_df и closed_eyes_df
merged_df = pd.merge(open_eyes_df, closed_eyes_df, on='patient_id', suffixes=('_open', '_closed'))
sfreq_open = open_eyes_df['sfreq'].iloc[0]
sfreq_closed = closed_eyes_df['sfreq'].iloc[0]

results = []

for _, row in merged_df.iterrows():
    psd_open = compute_psd(merged_df[[col + '_open' for col in sum(channels.values(), [])]].rename(columns=lambda x: x.replace('_open', '')), sfreq_open, rhythms, channels)
    psd_closed = compute_psd(merged_df[[col + '_closed' for col in sum(channels.values(), [])]].rename(columns=lambda x: x.replace('_closed', '')), sfreq_closed, rhythms, channels)
    rhythm_diff = psd_open - psd_closed
    rhythm_diff['patient_id_base'] = row['patient_id']
    results.append(rhythm_diff)

# Итоговый DataFrame с различиями
results_df = pd.concat(results)

# Сохранение различий в CSV
results_df.to_csv('topological_differences.csv', index=False)

# Визуализация различий для каждой топологической группы
for rhythm, group_diff in results_df.groupby(level=0):  # По ритмам
    # Установка размеров автоматически
    num_channels = len(group_diff.columns)  # Количество групп (например, центральных, фронтальных и т.д.)
    num_patients = len(group_diff.T)  # Количество пациентов

    # Увеличиваем fig в зависимости от количества данных
    plt.figure(figsize=(max(12, num_channels * 2), max(8, num_patients * 0.5)))  # Автоувеличение размера

    group_diff.iloc[:, :-1].T.plot(kind='bar', alpha=0.7)
    plt.title(f'Topological Differences in {rhythm.capitalize()} Rhythm', fontsize=16)
    plt.ylabel('Difference in PSD', fontsize=14)
    plt.xlabel('Channel Groups', fontsize=14)
    plt.xticks(rotation=45, ha='right', fontsize=12)  # Поворот подписей на 45 градусов
    plt.legend(title='Patient ID', loc='upper right', bbox_to_anchor=(1.3, 1))
    plt.grid(alpha=0.5)
    plt.tight_layout()  # Автоматическая корректировка
    plt.savefig(f'topological_differences_{rhythm}.png', bbox_inches='tight')  # Учет всех подписей
    plt.close()
```
